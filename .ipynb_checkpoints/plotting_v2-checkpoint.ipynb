```python
#Color and style settings

w_color = "#FF4892"
w_ecolor = "#610B21"
b_color = "#FFD966"
b_ecolor = "#5C4400"
d_color = "#99D9FF"
d_ecolor = "#004065"


t_colors=[w_color, #winery
         b_color, #brewery
         d_color #distillery
         ]

te_colors=[w_ecolor, #winery
          b_ecolor, #brewery
          d_ecolor #distillery
          ]

fontdict={"fontsize": 14,
        "fontweight" : "bold",
        "verticalalignment": 'baseline'}
```

```python
csv_path="google_sample.csv"
```

```python
google_df=pd.read_csv(csv_path)
```

```python
type_state=google_df
type_state=type_state.iloc[:,[6,2,3,4,5]]
type_state.head()
type_state=type_state.rename(columns={"name":"counter"})
type_state["counter"]=1
type_state_pivot=type_state.pivot_table(index="state",columns="est",values="counter", aggfunc="sum")
type_state_df=type_state_pivot.reset_index()
```

```python
state_counts=type_state_pivot
state_counts=state_counts.reset_index()
state_counts.reset_index()
state_counts.columns
state_counts=state_counts.iloc[:,[0,1,2,3]]
state_counts["BDW_sum"]=state_counts["brewery"]+state_counts["distillery"]+state_counts["winery"]
state_counts=state_counts.reset_index()
state_counts
#state_counts_pivot=state_counts.pivot_table(index="index",columns="state",values="BDW_sum", aggfunc="sum")
#sc_bar=state_counts.plot.bar(stacked=True, color=t_colors, edgecolor="black", linewidth="1", 
#                                                                           aa=True, figsize=(20,5))

w_counts_state=state_counts.iloc[:,[1,4]]
b_counts_state=state_counts.iloc[:,[1,2]]
d_counts_state=state_counts.iloc[:,[1,3]]
```

```python
map_df=type_state
temp_join=map_df
temp_join.index.names=["state"]
```

```python
state_list=[]
for i in range(len(state_counts["state"])):
    state_list.append(state_counts.state[i])

list_states=pd.DataFrame(state_list)
list_states.reset_index()
list_states.columns
list_states=list_states.rename(index=str, columns={0:"state"})
list_states=list_states["state"]

map_df.head()
```

```python
count_est=type_state_df.iloc[:,[0,1,2,3]].reset_index()
count_est=count_est.drop(columns=["index"])
x=np.array(count_est["state"])
hw=np.array(count_est["winery"])
hb=np.array(count_est["brewery"])
hd=np.array(count_est["distillery"])

count_b=plt.figure(figsize=(20,6),dpi=192, facecolor="#dddddd", frameon=True)
plt.bar(np.array(x),width=.25,height=(hb+hw+hd), color=t_colors, joinstyle="round")
#ylim(0,max(hw)+max(hb)+max(hd))
plt.bar(x=x,height=hd, color=d_color, edgecolor=d_ecolor, bottom=hb+hw)
plt.bar(x=x,height=hb, color=b_color, edgecolor=b_ecolor, bottom=hw)
plt.bar(x=x,height=hw, color=w_color, edgecolor=w_ecolor)
plt.ylabel("Number in Google Sample",labelpad=2 )
plt.xlabel("State", labelpad=2)
plt.title("Total Google Sample by Establishment Type", size=24, weight="bold")
w_leg=mpatches.Patch(color=w_color, label="Winery")
b_leg=mpatches.Patch(color=b_color, label="Brewery")
d_leg=mpatches.Patch(color=d_color, label="Distillery")
plt.legend(handles=[w_leg,b_leg,d_leg], loc="best", fancybox=True)
plt.grid(b=True)

count_b.savefig("test.png")
```

```python
map_df=type_state
temp_join=map_df
temp_join.index.names=["state"]
```

```python
mean_lat=pd.DataFrame(temp_join.groupby(["state"])["lat"].mean()).reset_index()
mean_lon=pd.DataFrame(temp_join.groupby(["state"])["lon"].mean()).reset_index()
mean_lon_lat=pd.merge(mean_lat,mean_lon, on="state")

mean_lon_lat=mean_lon_lat.rename(columns={"lat":"mean_lat","lon":"mean_lon"})
mean_lon_lat=mean_lon_lat.set_index("state")
```

```python
new_map=map_df.set_index("state")

new_map=new_map.join(new_map,lsuffix="_new_map", rsuffix="_mean_lon_lat")
new_map.head()
map_plot=new_map.iloc[:,[0,1,2,3]]
map_plot=map_plot.rename(columns={"est_new_map":"est","counter_new_map":"counter","lon_new_map":"lon","lat_new_map":"lat"})
map_plot=map_plot.reset_index()
```

```python
map_group=map_plot.reset_index()
map_group=pd.DataFrame(map_group.groupby(["state","est","lat","lon"])["counter"].sum()).reset_index()
map_group=map_group.drop_duplicates(subset=["state","est","counter"],keep="last")
map_group.head()
map_group.to_csv("map.csv")
```

```python
w_mapping=map_group.loc[map_group["est"]=="winery",:]
b_mapping=map_group.loc[map_group["est"]=="brewery",:]
d_mapping=map_group.loc[map_group["est"]=="distillery",:]

w_mapping=w_mapping.merge(w_counts_state, how="outer",on="state")
w_mapping=w_mapping.iloc[:,[0,1,2,3,5]]
w_mapping=w_mapping.rename(columns={"winery":"counts"})

b_mapping=b_mapping.merge(b_counts_state, how="outer",on="state")
b_mapping=b_mapping.iloc[:,[0,1,2,3,5]]
b_mapping=w_mapping.rename(columns={"brewery":"counts"})

d_mapping=d_mapping.merge(d_counts_state, how="outer",on="state")
d_mapping=d_mapping.iloc[:,[0,1,2,3,5]]
d_mapping=w_mapping.rename(columns={"distillery":"counts"})
```

```python
#Continential US and Hawaii mapping from http://boundingbox.klokantech.com:
#westlimit=-162.27; southlimit=16.31; eastlimit=-66.12; northlimit=49.11
#Continental US only from http://boundingbox.klokantech.com:
#westlimit=-126.2; southlimit=24.7; eastlimit=-65.8; northlimit=49.7
c_usa_lat = 49.7
c_usa_lon = -126.2
c_usa_llcnrlon = -126.2
c_usa_llcnrlat = 24.7
c_usa_urcrnrlon = -65.8
c_usa_urcrnrlat = 49.7

usa_lat = 49.11
usa_lon = -162.27
usa_llcnrlon = -162.27
usa_llcnrlat = 16.31
usa_urcrnrlon = -66.12
usa_urcrnrlat = 49.11
```

```python
wlon = np.array(w_mapping["lon"])
wlat = np.array(w_mapping["lat"])
wsize = np.array(w_mapping["counts"])
wlons = w_mapping["lon"].values
wlats = w_mapping["lat"].values

dlon = np.array(d_mapping["lon"])
dlat = np.array(d_mapping["lat"])
dsize = np.array(d_mapping["counts"])

blon = np.array(b_mapping["lon"])
blat = np.array(b_mapping["lat"])
bsize = np.array(b_mapping["counts"])

x=tuple(map_group["lon"])
y=tuple(map_group["lat"])
size=tuple(map_group["counter"].values)
```

```python
test_plot =plt.subplots(figsize=(20,10),dpi=192,sharex=True, sharey=True)
fig, ax = test_plot
ax.set_xlim([c_usa_lon,c_usa_urcrnrlon])   #longitude
ax.set_ylim([c_usa_llcnrlat,c_usa_lat])   #latitude
test_map = ax.scatter(x, y, s=dsize+bsize+wsize, color=t_colors, edgecolor="black", linewidth=.85, alpha=.75)
ax.scatter(x=wlon, y=wlat, s=wsize, color=w_color, edgecolor=w_ecolor, linewidths=1)
ax.scatter(x=blon, y=blat, s=bsize, color=b_color, edgecolor=b_ecolor, linewidths=1)
ax.scatter(x=dlon, y=dlat, s=dsize, color=d_color, edgecolor=w_ecolor, linewidths=1)
```

```python
map_plot=plt.subplots(figsize=(20,10))#, dpi=192)
fig,ax=map_plot
axw=ax
axd=ax
axb=ax
ax.set_ylim(min(map_df["lon"]),max(map_df["lon"]))   #longitude
ax.set_xlim(min(map_df["lat"]),max(map_df["lat"]))   #latitude
ax.map = Basemap(resolution='c', # c, l, i, h, f or None
            projection='merc',
            lat_0=c_usa_lat, lon_0=c_usa_lon,
            llcrnrlon=c_usa_llcnrlon, llcrnrlat= c_usa_llcnrlat, urcrnrlon=c_usa_urcrnrlon, urcrnrlat=c_usa_urcrnrlat)

ax.map.scatter(x, y, s=map_group["counter"], color=t_colors, edgecolor="black", latlon=True, linewidth=.85, alpha=.75, zorder=2)
axw.map.scatter(wlon, wlat, s=map_group["counter"].loc[map_group["est"]=="winery"]*45*(sum(map_group["counter"].loc[map_group["est"]=="winery"])/sum(map_group["counter"])), color=w_color, edgecolor=w_ecolor, latlon=True, linewidth=.85, alpha=.95, zorder=2)
axb.map.scatter(blon, blat, s=map_group["counter"].loc[map_group["est"]=="brewery"]*40*(sum(map_group["counter"].loc[map_group["est"]=="brewery"])/sum(map_group["counter"])), color=b_color, edgecolor=b_ecolor, latlon=True, linewidth=.85, alpha=.95, zorder=2)
axd.map.scatter(dlon, dlat, s=map_group["counter"].loc[map_group["est"]=="distillery"]*25*(sum(map_group["counter"].loc[map_group["est"]=="distillery"])/sum(map_group["counter"])), color=d_color, edgecolor=d_ecolor, latlon=True, linewidth=.85, alpha=.5, zorder=3)

ax.map.fillcontinents(color="#F2F2F2",lake_color="#46BCEC", zorder=1)
ax.map.drawmapboundary(fill_color="#46BCEC", zorder=0)
ax.map.drawcoastlines(2)
ax.map.readshapefile("us_shapefiles/us_shapefile", 
                     "STUSPS", 
                     drawbounds=True, 
                     zorder=None, 
                     linewidth=0.5, 
                     antialiased=10, 
                     default_encoding='utf-8')

plt.show()
#plt.savefig("map.png")
```
