# Data Analysis & Visualization - SDD Input

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import linregress
%matplotlib inline
```

```python
#Set the path for the data and read into a dataframe
data_path = "all_data.xls"
all_data_df = pd.read_excel(data_path)
```

## Everything below should be cut and paste into a master file

## Comparison of Google Maps vs. other data sources

```python
breweries_df = all_data_df[['State', 'brewery_count', 'google_brewery_count']].copy()
wineries_df = all_data_df[['State', 'winery_count', 'google_winery_count']].copy()
distilleries_df = all_data_df[['State', 'craft_distillery_count', 'google_distillery_count']].copy()
```

```python
#create a line that is y = x
line = np.arange(0,(max(breweries_df['brewery_count'] + 1)))
line_df = pd.DataFrame(line, columns = ['x_axis'])
line_df['y_axis'] = line
```

```python
#create a graph that gives the beermapping.com data vs. the google data
plt.plot(line_df['x_axis'], line_df['y_axis'], label = "Expected distribution")
plt.scatter(breweries_df["brewery_count"], breweries_df["google_brewery_count"], label = "Actual distribution")
plt.xlim(0,314)
plt.ylim(0, 315)
plt.xlabel("Beermapping.com Data")
plt.ylabel("Google Maps Data")
plt.title("USA Brewery Count: Google Maps Data vs. Beermapping.com Data")
plt.legend()
plt.savefig('brewery_comparison.png')
```

```python
#create a graph that gives the data.gov data vs. the google data
plt.plot(line_df['x_axis'], line_df['y_axis'], label = "Expected distribution")
plt.scatter(wineries_df["winery_count"], wineries_df["google_winery_count"], label = "Actual distribution")
plt.xlim(0,250)
plt.ylim(0, 250)
plt.xlabel("Data.gov Data")
plt.ylabel("Google Maps Data")
plt.title("USA Winery Count: Google Maps Data vs. Data.gov Data")
plt.legend()
plt.savefig('winery_comparison.png')
```

```python
#create a graph that gives the data.gov data vs. the google data
plt.plot(line_df['x_axis'], line_df['y_axis'], label = "Expected distribution")
plt.scatter(distilleries_df["craft_distillery_count"], distilleries_df["google_distillery_count"], label = "Actual distribution")
plt.xlim(0,120)
plt.ylim(0, 120)
plt.xlabel("Data.gov Data")
plt.ylabel("Google Maps Data")
plt.title("USA Distillery Count: Google Maps Data vs. Data.gov Data")
plt.legend()
plt.savefig('distillery_comparison.png')
```

## Correlation Analysis

```python
dependent_variables = ['brewery_count', 'winery_count', 'craft_distillery_count']
education_variables = ['birth_ed_less_than_high', 'birth_ed_high', 
                       'birth_ed_some_coll', 'birth_ed_batchelors', 
                       'birth_ed_graduate']
income_variables = ['birth_income_none', 'birth_income_less_10K', 'birth_income_10_15K',
                    'birth_income_15_25K', 'birth_income_25_35K', 'birth_income_35_50K', 
                    'birth_income_50_65K', 'birth_income_65_75K', 'birth_income_more_75K']
industry_variables = ['industry_ag_forestry_fish', 'industry_construction', 'industry_manufact',
                      'industry_wholesale', 'industry_retail', 'industry_trans', 'industry_information',
                      'industry_finance_insure_realestate', 'industry_professional_scientific_management',
                      'industry_education_health_care', 'industry_arts_entertatinment', 'industry_other_services',
                      'industry_public_admin', 'industry_military']
occupation_variables = ['occupation_manage_bus_science_arts', 'occupation_service', 'occupation_sales_office',
                        'occupation_resources_construct_maint', 'occupation_production_transport', 'occupation_military']
```

```python
#perform a correlation analysis and export to csv
correlation = all_data_df.corr()
correlation = correlation[dependent_variables] 
correlation.to_excel('correlation.xlsx', sheet_name = "data")
correlation.head()
```

```python
#Create a heatmap for the correlation between count of establishment and level of education by birth place
sns.heatmap(correlation.loc[education_variables])
plt.title("Heat Map Demonstrating Correlation between \n Count of Establishment and Education Level by Birth Place")
plt.savefig('correlation_education.png')
```

```python
#Create a heatmap for the correlation between count of establishment and income level
sns.heatmap(correlation.loc[income_variables])
plt.title("Heat Map Demonstrating Correlation between \n Count of Establishment and Income Level by Birth Place")
plt.savefig('correlation_income.png')
```

```python
#Create a heatmap for the correlation between count of establishment and industry employment
sns.heatmap(correlation.loc[industry_variables])
plt.title("Heat Map Demonstrating Correlation between \n Count of Establishment and Industry Employment")
plt.savefig('correlation_industry.png')
```

```python
#Create a heatmap for the correlation between count of establishment and occupational employment
sns.heatmap(correlation.loc[occupation_variables])
plt.title("Heat Map Demonstrating Correlation between \n Count of Establishment and Occupational Employment")
plt.savefig('correlation_occupation.png')
```
